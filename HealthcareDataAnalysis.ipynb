#**Data Analysis and Visualization**

## **Gender Distribution of Patients**

```python
gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(8,6))
gender_counts.plot.pie(autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightcoral'])
plt.title('Gender Distribution of Patients')
plt.ylabel('')
plt.show()
```

*The above graph on Gender Distribution of Patients, we can see that their are equal number of male and female patients, both having 50% distribution.*

## **5 Most Common Medicial Conditions**

```python
plt.figure(figsize=(10,6))
sns.countplot(y='Medical Condition', data=df, palette='viridis')
plt.title('Top 5 Most Common Medical Conditions')
plt.xlabel('Number of Patients')
plt.ylabel('Medical Condition')
plt.grid(axis='x', linewidth=0.5,color='black')
plt.show()
```

*The above graph represents the top 5 commom medicial conditions, each of 5 medical conditions have 8000+ patients. A little deep obeseravation into graph we can see that Diabetes and Arthritis have highest number of patients.*

## **Admissions per month**

```python
#Convert object datatype to datetime datatype
df['Date of Admission']=pd.to_datetime(df['Date of Admission'])
print("Converted to datetime datatype successfully...")
```

```python
#creating new column named as 'Month' from Date of Admission
df['Month']=df['Date of Admission'].dt.month

#Plotting
monthly_addmission=df['Month'].value_counts().sort_index()
plt.figure(figsize=(10,6))
monthly_addmission.plot(kind='bar', color='skyblue')
plt.title('Admissions per Month')
plt.xlabel('Month')
plt.ylabel('Number of Admissions')
plt.xticks(rotation=0)
plt.grid(color='blue', linestyle=':', linewidth=0.5)
plt.show()
```

*This graph represents the admission of patients per month, seeing this graph we can say that in every month there are 4000+ admissions. And month number 7 and 8 i.e. July and August have highest admission whereas month number 2 i.e. February has least number of admissions.*

##**Patients Admission Type**

```python
# Count the occurrences of each admission type
admission_counts = df['Admission Type'].value_counts()

# Plotting the donut chart
plt.figure(figsize=(8, 8))
plt.pie(admission_counts, labels=admission_counts.index, autopct='%1.1f%%', startangle=90,
        wedgeprops={'edgecolor': 'black', 'linewidth': 1, 'linestyle': 'solid', 'width': 0.4}, colors=['#C49C94', '#90EE90', '#ADD8FF'])

# Adding the title
plt.title('Patient Admission Type Distribution')

# Display the plot
plt.show()
```

*This graph give a pictorial idea of the admission type of the patients. Seeing the numerical data, we can say that Urgent Admission Type has the highest percentage whereas Emergency Admission Type has lowest percentage.*

##**Highest Age Group among Patients**

```python
# Define age bins and labels for age groups
age_bins = [0, 20, 40, 60, 80, 100]
age_labels = ['0-20', '21-40', '41-60', '61-80', '81+']

# Create a new column 'Age Group' based on the defined bins
df['Age Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)

# Plotting the histogram for age groups
plt.figure(figsize=(10, 6))
df['Age Group'].value_counts().sort_index().plot(kind='bar', color='lightgreen')

# Adding title and labels
plt.title('Distribution of Patients by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Patients')
plt.grid(color='blue', linestyle=':', linewidth=0.5)
# Display the plot
plt.show()
```

**This graphs shows the age distribution of patients.
There are 16000+ patients from the age group 21-80.
Highest number of patients are in the age group 41-60.
Lowest number of patients are in the age group 0-20.*

##**Line chart for blood groups**

```python
# Count the occurrences of each blood type
blood_type_counts = df['Blood Type'].value_counts().sort_index()

# Plotting the line chart
plt.figure(figsize=(10, 6))
plt.plot(blood_type_counts.index, blood_type_counts.values, marker='o', color='skyblue', linestyle='-', linewidth=2)

# Adding title and labels
plt.title('Distribution of Blood Types among Patients')
plt.xlabel('Blood Type')
plt.ylabel('Number of Patients')

# Display the plot
plt.show()
```

*This graph shows the blood distribution among patients. We can say that highest number of patients have blood group A- and lowest number of patients have blood group O-. Lastly there are almost equal number of patients having blood group AB+, AB-, B+ and B-.*

##**Distribution of Medications**

```python
# Count the occurrences of each medication
medication_counts = df['Medication'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(medication_counts, labels=medication_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen', 'orange'])

# Adding the title
plt.title('Distribution of Medication among Patients')

# Display the plot
plt.show()
```

*This graph represenst the Distribution of Medication among patients. The Lipitor has the higest distribution whereas Penicillin and Paracetamol has lowest distribution.
Penicillin and Paraetamol has equal distribution i.e. 19.9%.
Ibuprofen and Aspirin has equal distribution i.e. 20.0%.*

## **Distribution of Test Results**

```python
# Analyzing Test Results
test_results_count = df['Test Results'].value_counts()
print(test_results_count)

# Plotting Bar Graph for Test Results Distribution
plt.figure(figsize=(8, 5))
sns.barplot(x=test_results_count.index, y=test_results_count.values)
plt.title('Distribution of Test Results')
plt.xlabel('Test Result')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

*This graph represents the patients with Test results. Approximately, Abnormal, Normal and Inconclusive have equal number of counts, but Abnormal Test Result count is little higher than Normal Test Result.*

##**Insurance Provider Distribution**

```python
plt.figure(figsize=(8, 8))
insurance_counts = df['Insurance Provider'].value_counts()
insurance_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribution of Patients by Insurance Provider')
plt.ylabel('')
plt.show()
```

*This graph represents the distribution of patients by insurance provider. Seeing the chart, we can say that Cigna Insurance Provider has highest distribution of 20.3% whereas Aetna Insurance Provider has lowest distribution of 19.7%.*

##**Billing amount and Medical condition**

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Medical Condition', y='Billing Amount', data=df)
plt.title('Billing Amount by Medical Condition')
plt.xlabel('Medical Condition')
plt.ylabel('Billing Amount')
plt.xticks(rotation=45)
plt.grid(color='blue', linestyle=':', linewidth=0.5)
plt.show()
```

*This graph represents the Billing amount of Medical Condition. It is clearly visible that each medical condition has billing amount more than 50000, Hypertension condition has the highest Billing amount when compared to other conditions.*
